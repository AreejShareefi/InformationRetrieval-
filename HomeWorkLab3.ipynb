```python
import requests 

from bs4 import BeautifulSoup as bs 

#request the page & present its html code

r = requests.get('https://www.timeout.com/film/the-best-films-of-2022-so-far') 



soup = bs(r.content) 


print(soup.prettify()) 
```

```python
# using regular expression to print names of movies:- 

import re 

listOfMovies = soup.find_all('h3')

for tag in listOfMovies:
    movies=tag.contents[1]
    print(movies)
```
